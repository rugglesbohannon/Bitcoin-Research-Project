```python
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
import plotly.io as pio
from creds import *
import seaborn as sns
import numpy as np
import re
from plotly.subplots import make_subplots
px.set_mapbox_access_token(mapbox_access_token)
```

```python
#Improt daily stock price dataset in recent year
btc = pd.read_csv('BTC-USD.csv')
btc_daily = pd.read_csv('BTC-USD-Daily.csv')
BMW = pd.read_csv('BMW.DE.csv')
TSLA = pd.read_csv('TSLA.csv')
MSFT = pd.read_csv('MSFT.csv')
AT = pd.read_csv('T.csv')
NDSDAQ = pd.read_csv('NASDAQ.csv')
btc_daily.tail()
```

```python
#Merge the high price of each stocks in one dataset
btc_stock = pd.DataFrame(btc_daily, columns = ['High'])
btc_stock = btc_stock.rename(columns={"High": "BTC"})
btc_stock['BMW'] = BMW["High"]
btc_stock['TSLA'] = TSLA["High"]
btc_stock['MSFT'] = MSFT["High"]
btc_stock['AT'] = AT["High"]
btc_stock['NDSDAQ'] = NDSDAQ["High"]
#Test between stocks priece with Pearson correlation coefficient
btc_stock_corr = btc_stock.corr(method ='pearson')
btc_stock_corr.head()
```

```python
#Import monthly average price of currenies in recent five years
btc_monthly = pd.read_csv('BTC-USD-Monthly.csv')
CNY = pd.read_csv('CNYUSD.csv')
EUR = pd.read_csv('EURUSD.csv')
Gold = pd.read_csv('GC=F.csv')
Silver = pd.read_csv('SI=F.csv')
Palladium = pd.read_csv('PA=F.csv')
Platinum = pd.read_csv('PL=F.csv')
Gold.head()
```

```python
#Merge the high price of each currency in one dataset, and calculate the Pearson correlation coefficient
btc_currency = pd.DataFrame(btc_monthly, columns = ['High'])
btc_currency = btc_currency.rename(columns={"High": "BTC"})
btc_currency['CNY'] = CNY["High"]
btc_currency['EUR'] = EUR["High"]
btc_currency['Gold'] = Gold["High"]
btc_currency['Silver'] = Silver["High"]
btc_currency['Palladium'] = Palladium["High"]
btc_currency['Platinum'] = Platinum["High"]
btc_currency_corr = btc_currency.corr(method ='pearson')
btc_currency_corr.head()
```

```python
#Correlation matrix between bitcoin stocks and other tech companies stock price
mask = np.triu(np.ones_like(btc_stock_corr, dtype=bool)) #Masked the part that repeated 
rLT = btc_stock_corr.mask(mask)

#Heatmap of correlation matrix 
heat = go.Heatmap(
    z = rLT,
    x = rLT.columns.values,
    y = rLT.columns.values,
    zmin = - 0.25, # Sets the lower bound of the color domain
    zmax = 1,
    xgap = 1, # Sets the horizontal gap (in pixels) between bricks
    ygap = 1,
    colorscale = 'RdBu'
)

title = 'Bitcoin Stock Correlation Matrix'

#Update layout
layout = go.Layout(
    title_text=title, 
    title_x=0.5, 
    width=600, 
    height=600,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    yaxis_autorange='reversed'
)

fig=go.Figure(data=[heat], layout=layout)
fig.show()
pio.write_html(fig,"Bitcoin Stock Correlation Matrix.html", full_html=False)
```

```python
#orrelation matrix between bitcoin stocks and other currency and mental price 
mask = np.triu(np.ones_like(btc_currency_corr, dtype=bool))
rLT = btc_currency_corr.mask(mask)

heat = go.Heatmap(
    z = rLT,
    x = rLT.columns.values,
    y = rLT.columns.values,
    zmin = - 0.25, # Sets the lower bound of the color domain
    zmax = 1,
    xgap = 1, # Sets the horizontal gap (in pixels) between bricks
    ygap = 1,
    colorscale = 'RdBu'
)

title = 'Bitcoin Currency Correlation Matrix'

layout = go.Layout(
    title_text=title, 
    title_x=0.5, 
    width=600, 
    height=600,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    yaxis_autorange='reversed'
)

fig=go.Figure(data=[heat], layout=layout)
fig.show()
pio.write_html(fig,"Bitcoin Currency Correlation Matrix.html", full_html=False)
```

```python

for i, row in enumerate(btc["Date"]):
    p = re.compile(" 00:00:00")
    datetime = p.split(btc["Date"][i])[0]
    btc.iloc[i, 1] = datetime

fig = make_subplots(
    rows=3, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.03,
    specs=[[{"type": "table"}],
           [{"type": "scatter"}],
           [{"type": "scatter"}]]
)

fig.add_trace(
    go.Scatter(
        x=btc["Date"],
        y=btc["Close"],
        mode="lines",
        name="Bitcoin Close"
    ),
    row=3, col=1
)

fig.add_trace(
    go.Scatter(
        x=btc["Date"],
        y=btc["Volume"],
        mode="lines",
        name="Bitcoin Volume"
    ),
    row=2, col=1
)

fig.add_trace(
    go.Table(
        header=dict(
            values=["Date","Open","High","Low","Close","Adj Close"],
            font=dict(size=10),
            align="left"
        ),
        cells=dict(
            values=[btc[k].tolist() for k in btc.columns[1:]],
            align = "left")
    ),
    row=1, col=1
)
fig.update_layout(
    height=800,
    showlegend=False,
    title_text="Bitcoin Histrocial Data",
)

fig.show()
pio.write_html(fig,"Bitcoin Historical Data.html", full_html=False)
```
